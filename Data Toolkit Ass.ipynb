# **Difference between np.array(), np.asarray(), and np.asanyarray()**

# 1. np.array():
   
   This function creates new array object from any given input data like  list or tuple.
   
   Example:
     
     import numpy as np
     a = [7, 8, 9]
     array = np.array(a)
     

# 2. *np.asarray()*:
   
   This function convert input data into NumPy array but do not copy the data if it is already a NumPy array.
   
   Example:
     
     a = np.array([7, 8, 9])
     array = np.asarray(a)
     

# 3. *np.asanyarray()*:
   
  Same to np.asarray(), but it has more general, then it convert the input data into an array.
   
   Example:

     python
     a = np.matrix([7, 8], [5, 6])
     array = np.asanyarray(a)
     

### Deep Copy vs Shallow Copy

1. *Shallow Copy*:

   A shallow copy of an object creates  new object, but it copies the reference to objects containe within original object.

   Example:
     
     import numpy as np
     a = np.array([7, 8, 9])
     shallow_copy = a.view()
     

2. *Deep Copy*:

   A deep copy creates new object all objects found within the original object. Changes to the deep copy do not affect the original and vice versa.

     Example:
    
     a = np.array([7, 8, 9])
     deep_copy = np.copy(a)

# **Ans 14:-**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': pd.date_range(start='2024-09-01', periods=200, freq='D'),
    'Temperature': np.random.uniform(30, 55, size=200),
    'Humidity': np.random.uniform(60, 65, size=200)
}

df = pd.DataFrame(data)

fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:blue')
ax1.plot(df['Date'], df['Temperature'], color='tab:blue', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:red')
ax2.plot(df['Date'], df['Humidity'], color='tab:red', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Temperature and Humidity Over Time')

fig.tight_layout()
plt.show()
```

# **Ans 17:-**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

x = np.random.randn(200)
y = np.random.randn(200)

def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

quadrants = [get_quadrant(xi, yi) for xi, yi in zip(x, y)]

df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

plt.figure(figsize=(4, 9))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1', s=200)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Quadrant Y Scatter Plot')

plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

plt.legend(title='Quadrants')

plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 7 * np.pi, 300)
y = np.sin(x)

p = figure(title="sine wave function",
           x_axis_label='X-axis (radians)',
           y_axis_label='Y-axis (sin(x))',
           plot_width=800, plot_height=100)

p.line(x, y, legend_label="sin(x)", line_width=2, color="green")

p.xgrid.visible = True
p.ygrid.visible = True

show(p)
```

# **Ans 19:-**

```python
import random
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.transform import factor_cmap

output_notebook()

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(10, 100) for _ in categories]

source = ColumnDataSource(data=dict(categories=categories, values=values))

p = figure(x_range=categories, plot_height=200, plot_width=400, title="Random Catalogical Bar Chart",
           toolbar_location=None, tools="")

p.vbar(x='categories', top='values', width=0.9, source=source,
       fill_color=factor_cmap('categories', palette="Spectral5", factors=categories))

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
p.title.text = "Random Catalogical Bar Chart"

show(p)
```
